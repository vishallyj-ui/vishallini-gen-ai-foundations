```python
#Text Selection
df = pd.read_csv('Reviews.csv', nrows=50000)
print(" Dataset loaded successfully!")
print(f"Shape: {df.shape}")
print(df[['Text', 'Score']].head(3))
```

```python
#Vectorization Engine
sentence_1 = "This olive oil is wonderful and very light on the stomach."
sentence_2 = "The packaging was broken and let too much light inside the box."
sentence_3 = "I prefer this brand because it is a light and healthy snack."
sentences = [sentence_1, sentence_2, sentence_3]
print(" Sentences:")
for i, s in enumerate(sentences, 1):
    print(f"  Sentence {i}: {s}")
```

```python
#Vector Slicing
vectorizer = CountVectorizer()
dtm = vectorizer.fit_transform(sentences)
dense_matrix = dtm.toarray()
print(" Vocabulary:")
print(vectorizer.get_feature_names_out())
print()
print(" Document-Term Matrix (dense NumPy array):")
print(dense_matrix)
print(f"\nMatrix shape: {dense_matrix.shape}  →  (3 sentences × {dense_matrix.shape[1]} unique words)")
```

```python
#Similarity Mathematics (Dot Product)
vec_1 = dense_matrix[0]
vec_2 = dense_matrix[1]
vec_3 = dense_matrix[2]
print("vec_1 → Sentence 1 (light = texture/calorie):")
print(vec_1)
print("\nvec_2 → Sentence 2 (light = illumination):")
print(vec_2)
print("\nvec_3 → Sentence 3 (light = calorie/weight):")
print(vec_3)
```

```python
similarity_1_2 = np.dot(vec_1, vec_2)
similarity_1_3 = np.dot(vec_1, vec_3)
print("=" * 55)
print("         DOT PRODUCT SIMILARITY SCORES")
print("=" * 55)
print(f"  S1 vs S2 (Different contexts — light meaning differs): {similarity_1_2}")
print(f"  S1 vs S3 (Similar contexts  — light means same thing): {similarity_1_3}")
print("=" * 55)
```

```python
#Final Summary & Analysis
print("\n========== FINAL RESULTS SUMMARY ==========\n")
print(f'Sentence 1: "{sentence_1}"')
print(f'Sentence 2: "{sentence_2}"')
print(f'Sentence 3: "{sentence_3}"')
print(f"\n Dot Product (S1 vs S2) [Different Context]: {similarity_1_2}")
print(f" Dot Product (S1 vs S3) [Same Context]:      {similarity_1_3}")
print("\n KEY INSIGHT:")
print("   BoW assigns identical weight to 'light' in all sentences.")
print("   It cannot distinguish 'light' (illumination) from")
print("   'light' (low-calorie) — proving the Polysemy Barrier.")
print("\n Static count vectors fail to capture word meaning in context.")
```

Theory and Reflection Questions:

1. The Vector Conflict:
When I ran the dot product, I noticed that Sentence 1 and Sentence  two got a non-zero similarity score even though "light" means completely different things in each. This is because CountVectorizer just counts how many times a word appears, it has no idea what the word actually means in that sentence. So whether "light" refers to calories or to actual light coming through a box, the model treats it as the exact same thing. That's the core problem with static vectors.

2. The Context Blindspot:
The biggest issue with Bag-of-Words is that it completely throws away the meaning behind words. Imagine searching "light snack" on a website a BoW-based search engine could easily return results about light bulbs just because the word "light" appears. The same problem hits chatbots, which could misunderstand what a user is actually asking. Once you convert text into word counts, all the context and real meaning is just gone.

3.The Gen-AI Architectural Fix:
This is exactly the problem that modern LLMs like GPT solve using Self-Attention. Instead of giving "light" one fixed vector forever, the model looks at all the surrounding words and builds a meaning based on context. So "light" next to "stomach" gets a totally different vector than "light" next to "box." That's how Context-Aware Embeddings work the same word gets a different mathematical identity depending on where it appears.
